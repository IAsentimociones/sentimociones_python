```python
import requests
from bs4 import BeautifulSoup

#URL = "https://www.letras.com/?q=pelele"
#URL = "https://realpython.github.io/fake-jobs/"
#URL = "https://www.musica.com/letras.asp?t2=pelele+morad""
URL = "https://www.lyrics.com/lyrics/pelele%20morad"

page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
print(soup)

```

```python
results = soup.find(id="content-body")
#results = soup.find(recursive="resultado")
#results = soup.find_all("table", class_="rst")
print(results.prettify())
#print(results)
```

```python
job_elements = results.find_all("div", class_="sec-lyric clearfix")
#job_elements = results.find("td")
print(job_elements)

for job_element in job_elements:
    print(job_element, end="\n"*2)
```
